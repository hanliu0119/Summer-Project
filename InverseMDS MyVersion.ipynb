# MDS Function

```python
import sklearn.metrics.pairwise
import sklearn.manifold

def cal_distance(dataHD):
    distHD = sklearn.metrics.pairwise.manhattan_distances(dataHD)
    distHD = pd.DataFrame(distHD, columns=dataHD.index, 
                          index=dataHD.index)
    # set up the parameters:
    mds = sklearn.manifold.MDS(n_components=2, dissimilarity='precomputed', n_init=100, max_iter=1000)

    # reduction algorithm happens here:
    data2D = mds.fit_transform(distHD)
    data2D = pd.DataFrame(data2D, columns=['x','y'], index=dataHD.index)
    return data2D
```

```python
def proposal(current, step, direction):
    #print("proposal---------------------------------------")
    output = - 1
    if direction == 0:
        output = current - 0.5*step
    elif direction == 1:
        output = current + 0.5*step
    if output >= 0 and output <= 1:
        output = current
    return output

def stress(highDD, lowDD, weight):
    highD = highDD.copy()
    lowD = lowDD.copy()
    #highD = dataHD.copy()

    #print(highD.head())
    #print("stress----------------------------------------")
    #print(lowD.head())
    #print(weight)
    row = len(highD)
    col = len(highD.columns.tolist())
    
    highDSum = 0
    output = 0
    
    tmp_highD = highD.copy()
    for i in range(len(tmp_highD.columns.tolist())):
        tmp_highD.iloc[:, i] = [x*weight[i] for x in tmp_highD.iloc[:, i]]
    #print("stress2----------------------------------------")
    #print(tmp_highD.head())
    new_lowD = cal_distance(tmp_highD)

    #print(new_lowD.shape, lowD.shape)
    # scale lowD sum to 2
    tmp = new_lowD.values.sum()/2
    new_lowD = new_lowD/tmp
    #print(new_lowD.head())
    #print(new_lowD.values.sum(), lowD.values.sum())
    #print(abs(new_lowD.values-lowD.values).sum())
    return abs(new_lowD.values-lowD.values).sum()

def invMDs(highDD, lowDD, weights):
    highD = highDD.copy()
    lowD = lowDD.copy()
    #print(highD.head())
    #print(lowD.head())
    row = len(highD)
    col = len(highD.columns.tolist())    
    flag = [0]*col

    MAX = 50
    
    dummy_weights = []
    for i in range(len(lowD.columns)):
        dummy_weights = dummy_weights + [1/len(lowD.columns)]
    #print(len(dummy_weights), len(lowD.columns))
    #print(dummy_weights)
    #print(lowD.head())
    for i in range(len(lowD.columns)):
        lowD.iloc[:, i] = [x*dummy_weights[i] for x in lowD.iloc[:, i]]
    
    # scale lowD sum to 2
    tmp = lowD.values.sum()/2
    lowD = lowD/tmp
    
    for column in highD.columns:
        highD[column] = (highD[column]-highD[column].mean())/highD[column].std()

    proposedMove = 0
    step = [0]*col
    curWeight = [0]*col
    newWeight = [0]*col

    for i in range(col):
        step[i] = 2.0/col
    
    curWeight = weights
    stressCur = 0
    streeNew = 0
    
    stressCur = stress(highD, lowD, curWeight)
    #print(stressCur)
    #print("1----------------------------------------------------------------")
    for i in range(MAX):
        print("i = ", i)
        direction = 1
        for dim in range(col):
            #print("2-------------------------------------------------------")
            print("round=", i,"  dim=", dim, "  flag[dim]=", flag[dim], "  step[dim]=", step[dim], "  dir=",direction)
            #print("currWeight = ", curWeight)
            newWeight = curWeight
            if flag[dim] >= 5:
                step[dim] = step[dim]*2
                flag[dim] = 0
            elif flag[dim] <= -5:
                step[dim] = step[dim]/2
                flag[dim] = 0
            
           
            newWeight[dim] = proposal(curWeight[dim], step[dim], direction)
            newWeight = [v/sum(newWeight) for v in newWeight]
            print("Current Weight=", curWeight[dim], "  New Weight=", newWeight[dim])
            stressNew = stress(highD, lowD, newWeight)
            #print("newWeight Sum = ", sum(newWeight))
            print("streenew = ", stressNew)
            print("streecur = ", stressCur)
            if stressNew < stressCur:
                curWeight = newWeight
                stressCur = stressNew
                flag[dim] = flag[dim] + 1
            else:
                flag[dim] = flag[dim] - 1
                # change the direction if current stress is less than new stress
                if direction == 0:
                    direction = 1
                elif direction == 1:
                    direction = 0
    print(curWeight)
    return curWeight
```
